```python
def arithmetic_arranger(problems, display_answer = False):
    prob_len = len(problems)
    if prob_len > 5:
        return("Error: Too many problems.")

    operand1 = []
    operator1 = [] 
    operand2 = []
    offset = []
    answerlist = []
    is_minus = False
    is_plus = False
    
    for element in problems:
        is_minus = False
        is_plus = False
        #print("Element is: ", element)
        element = element.split()
        #print(element)
        digitX = element[0]
        digitY = element[2]
        
        for x in digitY:
            if not x.isdigit():
                return("Error: Numbers must only contain digits.")
        for x in digitX:
            if not x.isdigit():
                return("Error: Numbers must only contain digits.")
            
        #if(not (element[0].isdigit or element[2].isdigit)):
            #return("Error: Numbers must only contain digits.")
        if(len(element[0]) > 4 or len(element[2]) > 4):
            return("Error: Numbers cannot be more than four digits.")
        else:
            operand1.append(element[0])
            operand2.append(element[2])
            
        if(element[1] == "+"):
            operator1.append(element[1])
            is_plus = True
        elif(element[1] == "-"):
            operator1.append(element[1])
            is_minus = True
        else:
            return("Error: Operator must be '+' or '-'.")   
    #print(operand1)
    #print(operator1)
    #print(operand2)
    
        space = "    "
        
        if len(element[0]) > len(element[2]):
            offset.append(len(element[0]))
        elif len(element[0]) < len(element[2]):
            offset.append(len(element[2]))
        else:
            offset.append(len(element[2]))
            
        if(is_plus):
            answer = int(element[0]) + int(element[2])
            answerlist.append(answer)
            
        elif(is_minus):
            answer = int(element[0]) - int(element[2])
            answerlist.append(answer)
            
    #print(operand1[0].rjust(offset[0]+2))
    #print(operator1[0])
    #print(operand2[0].rjust(offset[0]+2))
    #print((offset[0]+2) * "-")
    #print(str(answerlist[0]).rjust(offset[0]+2))
    answerthis =""
    for x in range(len(operand1)):
        if x == (len(operand1)-1):
            keep = operand1[x].rjust((offset[x]+2))
            answerthis+= keep
        else:
            keep = operand1[x].rjust((offset[x]+2))
        #print(keep +"   ",end=" ")
            answerthis+= keep +"   " +" "
    #print("")
    answerthis+="\n"
    for x in range(len(operand1)):
        if(x == 0):
            operator = operator1[x]
            #print(keep+"   ",end=" ")
            answerthis+=operator#+"   "+" "
            number = operand2[x].rjust((offset[x])+1)
            answerthis += number
        else:
            space = " " * 4
            operator = operator1[x]
            number = operand2[x].rjust((offset[x])+1)
            answerthis+=space+operator+number
    answerthis+="\n"
    
    #for x in range(len(operand1)):
        #keep = operand2[x].rjust((offset[x])+2)
        #print(keep+"   ",end=" ")
        #answerthis += keep +"   "+" "
    #print("")
    for x in range(len(operand1)):
        if x == (len(operand1)-1):
            keep = ((offset[x])+2) * "-"
            answerthis += keep
        else:
            keep = ((offset[x])+2) * "-"+"   "
            #print(keep,end=" ")
            answerthis += keep + " "
        
    #print("")
    
    if(display_answer):
        answerthis+="\n"
        for x in range(len(operand1)):
            keep = str(answerlist[x]).rjust((offset[x])+2)+"   "
            #print(keep,end=" ")
            if x == (len(operand1)-1):
                answerthis+=keep
            else:
                answerthis+=keep+" "  
        answerthis = answerthis[:len(answerthis)-3]
        return answerthis
    answerthis = answerthis[:len(answerthis)]
    return answerthis
```

```python
print(arithmetic_arranger(["32 + 698", "3801 - 2", "45 + 43", "123 + 49"]))
```
